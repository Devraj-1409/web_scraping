# Web Scraping Project: 2

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_countries_by_wealth_per_adult"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text,'html')
```

```python
table = soup.find_all('table')[1]
```

```python
table_headers_scrap = table.find_all('th')
print(table_headers_scrap)
```

```python
table_headers = [item.text.strip().replace('\xa0', '') for item in table_headers_scrap]
print(table_headers) #use "." to combine to fn..it ensures that each method is applied to the output of previous output
```

```python
import pandas as pd
df = pd.DataFrame(columns = table_headers)
df
```

```python
table_rows_scrap = table.find_all('tr')
print(table_rows_scrap)
```

```python
for item in table_rows_scrap[1:]:
    individual_data_scrap= item.find_all('td')
    individual_data = [item.text.strip() for item in individual_data_scrap]
    #print(individual_data)
    length = len(df)
    df.loc[length] = individual_data
df
```
